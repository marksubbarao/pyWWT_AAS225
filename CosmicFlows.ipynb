### CosmicFlows

Vizier catalog: J/AJ/146/86/table1

```python
Cats = v.get_catalogs('J/AJ/146/86')
CosmicFlows=Cats[1]
CosmicFlows.keep_columns(["GGLON","GGLAT","__Dist_","__PV_","__Bmag_","__Kmag_"])
CosmicFlows.rename_column('__Dist_', 'Dist')
CosmicFlows.rename_column('__PV_', 'PV')
CosmicFlows.rename_column('__Bmag_', 'Bmag')
CosmicFlows.rename_column('__Kmag_', 'Kmag')
```

```python
from astropy import coordinates, units
coordsCol=coordinates.Galactic(CosmicFlows['GGLAT'],CosmicFlows['GGLON'],unit=(units.degree, units.degree),distance=coordinates.Distance(CosmicFlows['Dist'],units.mpc))
CosmicFlows.add_column(Column(coordsCol.icrs.ra,name='RA',meta={'ucd': 'pos.eq.ra'}),0)
CosmicFlows.add_column(Column(coordsCol.icrs.dec,name='Dec',meta={'ucd': 'pos.eq.dec'}),1)
```

### Supernovae

```python
Cats = v.get_catalogs('J/AJ/135/1766/SNe')
SNeCat=Cats[0]
SNeCat.keep_columns(["SN","_RAJ2000","_DEJ2000","SNtype","zSN","ObsDate","Ep.p"])
SNeCat.rename_column('_RAJ2000', 'RA')
SNeCat.rename_column('_DEJ2000', 'dec')
```

```python
print SNeCat
```

```python
timeCol=Time(SNeCat['ObsDate'], format='isot', scale='utc')
EpCol = TimeDelta(SNeCat['Ep.p'],format='jd')
startTime=timeCol-EpCol
```

```python
#Add Time column in WWT format-
timeList=[]
for i in range(len(startTime)):
    timeStr=str(startTime[i])
    timeString= timeStr[5:7]+'/'+timeStr[8:10]+'/'+timeStr[0:4]+' '+timeStr[11:19]
    timeList.append(timeString)
SNeCat.add_column(Column(timeList,name='TimeAndDate'))
```

```python
print SNeCat
```

```python
#Set up WWT layer
SNe_layer = wwt.new_layer("Dynamic Universe", "Supernovae", SNeCat.colnames)
#Set visualization parameters in WWT
props_dict = {'CoordinatesType':'Spherical',\
              'MarkerScale':'Screen',\
              'PointScaleType':'Constant',\
              'ScaleFactor':'64',\
              'ShowFarSide':'True',\
              'RaUnits':'Degrees',\
              'PlotType':'Gaussian',\
              'ColorValue':'ARGBColor:255:255:255:255',\
              'TimeSeries':'False'}
SNe_layer.set_properties(props_dict)
#Send data to WWT client
SNe_layer.update(data=SNeCat, purge_all=True, no_purge=False, show=True)
```

```python
del SNeCat['ObsDate']
```

```python
SNe_layer.set_property('BeginRange','1/1/2008 2:04:12 AM')
SNe_layer.set_property('EndRange','12/31/2008 2:04:12 AM')
```

### Cosmological Box calculations

```python
redshifts = [0.02,0.075,0.2,0.4,0.6,1.0,2,4]
distCol=Column(Planck13.comoving_distance(redshifts),name='comoving_distance')
scalefactor = Column(Planck13.scale_factor(redshifts),name="a")
times = Column(Planck13.age(redshifts),name="T")
data=Table([redshifts,distCol,scalefactor,times])
```
